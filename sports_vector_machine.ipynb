```python
import codecademylib3_seaborn
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from svm_visualization import draw_boundary
from players import aaron_judge, jose_altuve, david_ortiz

fig, ax = plt.subplots()

# Exploring Data:
# print(aaron_judge.columns)

# Checking the different values in the description column:
print(aaron_judge.description.unique())

# Cheking for unique values in the type column:
print(aaron_judge.type.unique())

# Creating the labels:
def players(player_name):
  player_name['type'] = player_name['type'].map({'S':1, 'B':0})

  # print(player_name.type)

  # Looking at the ball's location:
  # print(player_name[['plate_x', 'plate_z']])

  # Removing all NaN values:
  player_name = player_name.dropna(subset = ['type', 'plate_x', 'plate_z'])

  # Plotting the pitches:
  plt.scatter(x = player_name.plate_x, y = player_name.plate_z, c = player_name.type, cmap = plt.cm.coolwarm, alpha = 0.5)

  # Building the SVM:
  training_set, validation_set = train_test_split(player_name, random_state = 1)

  # Finding the best gamma and C value:
  classifier = SVC(kernel = 'rbf', gamma = 3, C = 1)
  classifier.fit(training_set[['plate_x', 'plate_z']], training_set['type'])

  # Visualizing the SVM model:
  draw_boundary(ax, classifier)
  ax.set_ylim(-2, 6)
  ax.set_xlim(-3, 3)
  plt.show()

  # Optimizing the SVM model:
  print(classifier.score(validation_set[['plate_x', 'plate_z']], validation_set['type']))


# Calling the players function on aaron_judge:
players(aaron_judge)

# Calling the players function on jose_altuve:
players(jose_altuve)

# # Calling the players function on david_ortiz:
players(david_ortiz)
```
